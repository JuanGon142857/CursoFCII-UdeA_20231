```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Trayectoria circular

```python
# lee el archivo de texto
datos = pd.read_csv('../example/circular.dat', sep = " ")

# separa los datos en dos arreglos
x = datos['x']
y = datos['y']
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trayectoria circular')
plt.savefig('circular.png')
```

### Espiral

```python
datos_espiral=pd.read_csv('../example/espiral.dat', sep = " ")
x_espiral=datos_espiral['x']
y_espiral=datos_espiral['y']
z_espiral=datos_espiral['z']
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x_espiral, y_espiral, z_espiral, 'gray')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.savefig('espiral.png')
```
